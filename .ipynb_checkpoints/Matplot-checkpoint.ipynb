## Counter

```python
c.update(['C++', 'Python'])
c
```

```python
from collections import Counter

with open('data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    
    languge_counter = Counter()
    row = next(csv_reader)
    for row in csv_reader:
        languge_counter.update(row['LanguagesWorkedWith'].split(';'))
print (languge_counter)
print (languge_counter.most_common(15))
```

```python
languges = []
popularity = []

for item in languge_counter.most_common(15):
    languges.append(item[0])
    popularity.append(item[1])
print (languges)
print (popularity)
```

```python
plt.bar(languges, popularity)
plt.title('Most Popular Languages')
plt.xlabel('Programming Langauages')
plt.ylabel('Number of people Sse')
plt.tight_layout()
plt.show()
```

```python
plt.barh(languges, popularity)
plt.title('Most Popular Languages')
#plt.ylabel('Programming Langauages')
plt.xlabel('Number of people Sse')
plt.tight_layout()
plt.show()
```

```python
languges.reverse()
popularity.reverse()
plt.barh(languges, popularity)
plt.title('Most Popular Languages')
#plt.ylabel('Programming Langauages')
plt.xlabel('Number of people Sse')
plt.tight_layout()
plt.show()
```

## Filling Area of Plot

```python
data = pd.read_csv('data1.csv')
print (data.columns)
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

```

```python
plt.plot(ages, dev_salaries, color='#444444',
            linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287
plt.fill_between(ages, py_salaries, alpha=0.25)
plt.legend()
plt.title('Medain Salary(USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary(USD)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#444444',
            linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287
plt.fill_between(ages, py_salaries, overall_median,
                 interpolate=True, alpha=0.25)
plt.legend()
plt.title('Medain Salary(USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary(USD)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#444444',
            linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287
plt.fill_between(ages, py_salaries, overall_median,
                 where=(py_salaries > overall_median),
                 interpolate=True, alpha=0.25)
plt.legend()
plt.title('Medain Salary(USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary(USD)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#444444',
            linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287

plt.fill_between(ages, py_salaries, overall_median,
                 where=(py_salaries > overall_median),
                 interpolate=True, alpha=0.25)
plt.fill_between(ages, py_salaries, overall_median,
                 where=(py_salaries <= overall_median),
                 interpolate=True, color='red', alpha=0.25)
plt.legend()
plt.title('Medain Salary(USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary(USD)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#444444',
            linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287

plt.fill_between(ages, py_salaries, dev_salaries,
                 where=(py_salaries > dev_salaries),
                 interpolate=True, alpha=0.25, label='Above Avg')
plt.fill_between(ages, py_salaries, dev_salaries,
                 where=(py_salaries <= dev_salaries),
                 interpolate=True, color='red', alpha=0.25, label='Below Avg')
plt.legend()
plt.title('Medain Salary(USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary(USD)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
from itertools import count
from matplotlib.animation import FuncAnimation
from matplotlib import pyplot as plt
import random

#plt.style.use('fivethirtyeight')
x_vals = []
y_vals = []

#plt.plot(x_vals, y_vals)

index = count()
def animate(i):
    x_vals.append(next(index))
    print ('animate')
    y_vals.append(random.randint(0,5))
    plt.cla() ## Clear axes
    plt.plot(x_vals, y_vals)

ani = FuncAnimation(plt.gcf(), animate, interval=1000)

#data = pd.read_csv('data4.csv')

plt.tight_layout()
plt.show()
```

#### CSV Change Continuesly

```python
import csv 
import random
import time

x_value = 0
total_1 = 1000
total_2 = 1000

fieldnames = ['x_value', 'total_l', 'total_2']

with open('data22.csv', 'w')as csv_file:
    csv_writer = csv.DictWriter(csv_file, fieldnames = fieldnames)
    csv_writer.writeheader()
    
while True:
    
    with open('data22.csv', 'a')as csv_file:
        csv_writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
        
        info = {
            'x_value': x_value,
            'total_l': total_1,
            'total_2': total_2
        }

        csv_writer.writerow(info)
        print (x_vals, total_1, total_2)

        x_value += 1
        total_1 = total_1 + random.randint(-6, 8)
        total_2 = total_2 + random.randint(-5, 6)
    
    time.sleep(1)
    
    
```
